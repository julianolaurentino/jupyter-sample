2. Polimento
    *   Padronização, tratamento, tipo de dados 

```python
# Alterando o tipo de dado da coluna 'Date' para datetime
df['Date'] = pd.to_datetime(df['Date'], format = "mixed", dayfirst=True)
df.head(5)
```

```python
# Renomeando Colunas
df.rename(columns={'Campaign Name': 'campaign_name'}, inplace=True)
df.rename(columns={'Spend [USD]': 'spend'}, inplace=True)
df.rename(columns={'# of Impressions': 'impressions'}, inplace=True)
df.rename(columns={'Reach': 'reach'}, inplace=True)
df.rename(columns={'# of Website Clicks': 'clicks'}, inplace=True)
df.rename(columns={'# of Searches': 'searches'}, inplace=True)
df.rename(columns={'# of View Content': 'content'}, inplace=True)
df.rename(columns={'# of Add to Cart': 'add_cart'}, inplace=True)
df.rename(columns={'# of Purchase': 'purchase'}, inplace=True)
df.head(5)
```

```python
# Tratar linhas preenchendo valores vazios
df['spend'] = df['impressions'].fillna(0).astype(float)
df['impressions'] = df['impressions'].fillna(0).astype(float)
df['reach'] = df['reach'].fillna(0).astype(float)
df['clicks'] = df['clicks'].fillna(0).astype(float)
df['searches'] = df['searches'].fillna(0).astype(float)
df['content'] = df['content'].fillna(0).astype(float)
df['add_cart'] = df['add_cart'].fillna(0).astype(float)
df['purchase'] = df['purchase'].fillna(0).astype(float)
```

```python
# Padronização - deixar todos os nomes de campanhas em letras minúsculas
df['campaign_name'] = df['campaign_name'].str.lower()
df.info()
```

3. Exportar

```python
df.to_parquet(output, index=False)
```
