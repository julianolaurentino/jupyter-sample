```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv(r'C:\Users\Casal-Inked\OneDrive - sga.pucminas.br\Dataset\Google_Ads_Dataset.csv', sep=',')
```

```python
df.info()
```

```python
# print(df.loc[:'campaign_name'])
df['Campaign_Name'] = df['Campaign_Name'].astype('string')

df['Campaign_Name']
```

```python
substituicoes_campaign_name = {
    '@':'a',
    '0':'o',
    '_':'',
    '3':'e'
}

df['Campaign_Name'] = df['Campaign_Name'].replace(substituicoes_campaign_name, regex=True).str.lower().str.strip()
```

```python
df['Campaign_Name'].unique()
```

```python
df[['Cost', 'Sale_Amount']] = df[['Cost', 'Sale_Amount']].replace('[^0-9.]', '', regex=True).astype(float)
```

```python
df[['Cost','Conversions','Leads', 'Conversion Rate']] = df[['Cost','Conversions','Leads', 'Conversion Rate']].fillna(0)
```

```python
df['Leads'] = df['Leads'].astype(int)
df['Conversions'] = df['Conversions'].astype(int)
df['Ad_Date'] = pd.to_datetime(df['Ad_Date'], format = 'mixed', dayfirst=True)
df['Ad_Date']
```

```python
df['Location'] = df['Location'].str.lower().astype('string')
df['Device'] = df['Device'].str.lower().astype('string')
```
